```python
#  The relationship between gravitational force and two bodies
import matplotlib.pyplot as plt

# menampilkan graphic
def tampil_grhp(x, y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Jarak dengan satuan meter')
    plt.ylabel('Gravitasi force dengan satuan newtowns')
    plt.title('Gravitasi Force dan Jarak')
    plt.show();

def hasil_F_r():
    # Hasil nilai untuk r
    r = range(100, 1001, 50)
    # list kosong untuk mengirim hitungan nilai dari F
    F = []
    
    # Hitungan, G
    G = 6.674*(10**-11)
    # dua masa
    m1 = 0.5
    m2 = 1.5
    
    # Menghitung force dan menambah F untuk ke list
    for dist in r:
        force = G*(m1*m2)/(dist**2)
        F.append(force)
    # Memanggil fungsi tampil_grhp
    tampil_grhp(r, F)

if __name__ == '__main__':
    hasil_F_r()
print('Visualisasi antara gravitasi force dengan jarak')
```

```python
# Drawing the Trajectory
# Draw the Tracjectory of a body in projectile motion

from matplotlib import pyplot as plt
import math

def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('x-cordinate')
    plt.ylabel('y-cordinate')
    plt.title('Projectile motion of a ball')
    
def frange(start, final, interval):
    
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
        
    return numbers
def draw_tracjectory(u, theta):
    
    theta = math.radians(theta) # 1
    g = 9.8
    
    # Time of flight
    t_flight = 2*u*math.sin(theta)/g # 2
    # Find time intervals
    intervals = frange(0, t_flight, 0.001)
    # List of x and y cordinates
    x = []
    y = []
    for t in intervals: # 3
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*g*t*t)
    draw_graph(x, y)

if __name__ == '__main__':
    try: # 4
        u = float(input('Enter the initial velocity (m/s): '))
        theta = float(input('Enter the angle of projection (degrees): '))
    except ValueError:
        print("You entered an individual input")
    else:
        draw_tracjectory(u, theta)
        plt.show()
    print("The trajectory of a ball when thrown with a velocity of 25 m/s at an angle of 60 degrees")
```

## 3: Enhaced Projectile Trajectory Comparison Program



```python
'''
Compare the projectile motion of abody theown with various combination of
initial velocity and angle of projection

'''

import matplotlib.pyplot as plt
import math

g = 9.8

def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('x-cordinate')
    plt.ylabel('y-cordinate')
    plt.title('Projectile motion at diffrent initial velocites and angles')

def frange(start, final, interval):
    
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
        
    return numbers

def draw_trajectory(u, theta, t_flight):
    # List of x and y cordinates
    x = []
    y = []
    intervals = frange(0, t_flight, 0.001)
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*g*t*t)
        
    # create the graph
    draw_graph(x, y)

if __name__ == '__main__':
    
    num_trajectories = int(input('How many trajectories? '))
    
    velocities = []
    angles = []
    
    for i in range(1, num_trajectories+1):
        v = input('Enter the initial velocity for trajectory {0} (m/s): '.format(i))
        theta = input('Enter the angle of projections for trajectory {0} (degrees): '.format(i))
        velocities.append(float(v))
        angles.append(math.radians(float(theta)))
        
    for i in range(num_trajectories):
        # Caculate time of flight, maximum horizontal distance and maximum vertical distance
        t_flight = 2*velocities[i]*math.sin(angles[i])/g
        S_x = velocities[i]*math.cos(angles[i])*t_flight
        S_y = velocities[i]*math.sin(angles[i])*(t_flight/2) - (1/2)*g*(t_flight/2)**2
        print('Initial velocity: {0} Angle of Projection: {1}'.format(velocities[1],
                                                                     math.degrees(angles[i])))
        print('T: {0} S_x: {1} S_y: {2}'.format(t_flight, S_x, S_y))
        print()
        draw_trajectory(velocities[i], angles[i], t_flight)
        
    # Add a legend and show the graph
    legends = []
    for i in range(0, num_trajectories):
        legends.append('{0} - {1}'.format(velocities[i], math.degrees(angles[i])))
    plt.legend(legends)
    plt.show()
    
```

```python
# Visualizing weekly expenditure using a bar chart

import matplotlib.pyplot as plt

def create_bar_chart(data, labels):
    # Num of bars
    num_bars = len(data)
    # This list is the point on the y-axis where each bar is centered. here it will be [1,2,3..]
    positions = range(1, num_bars+1)
    plt.barh(positions, data, align='center')
    plt.yticks(positions, labels)
    plt.xlabel('Amount')
    plt.ylabel('Categories')
    plt.title('Weekly expenditures')
    # Truns on the grid which may assist in visual estimation
    plt.grid()
    plt.show()
    
if __name__ == '__main__':
    n = int(input('Enter the number of categories: '))
    labels = []
    expenditures = []
    for i in range(n):
        category = input('Enter category: ')
        expenditure = float(input('Expenditure: '))
        
        labels.append(category)
        expenditures.append(expenditure)
    create_bar_chart(expenditures, labels)
```

## 5: Exploring the Relationship Between the Fibonacci Squence and the golden ratio





```python
# Relationship between Fibonacci squence and golden ratio

import matplotlib.pyplot as plt

def fibo(n):
    if n == 1:
        return [1]
    if n == 2:
        return [1, 1]
    # n > 2
    a = 1
    b = 1
    # First two members of the series
    series = [a, b]
    for i in range(n):
        c = a + b
        series.append(c)
        a = b
        b = c
    return series

def plot_ratio(series):
    ratios = []
    for i in range(len(series)-1):
        ratios.append(series[i+1]/series[i])
    plt.plot(ratios)
    plt.title('Ratio between Fibonacci numbers & Golden Ratio')
    plt.ylabel('Ratio')
    plt.xlabel('N0.')
    plt.show()

if __name__ == '__main__':
    # Number of fibonacci numbers
    num = 100
    series = fibo(num)
    plot_ratio(series)
    print(series)
```
